## Experiment 1

```python
import os
import subprocess
import tempfile
import pickle
from itertools import product
import torch.random
import timm
from src.utils.Config import Config
# from src.compute_cossim_dots_hierarchical import generate_dataset_rndset
from src.experiment_1.distance_helper import compute_distance_set
from itertools import product
from src.utils.create_stimuli.drawing_utils import *
from src.utils.misc import *
from src.utils.misc import main_text_nets, all_nets, appendix_nets, brain_score_nn, additional_nets

# 1) build your parameter grid
pretraining     = ['ImageNet']
background      = ['random']       # or ['black','white','random']
distance_types  = ['euclidean','cossim']
nets            =  additional_nets #['coatnet_0_rw_224'] #list(brain_score_nn.keys())  # list of nets
batch_size = 16

all_exps = product(pretraining, nets, background, distance_types)
arguments = [
    dict(
        pretraining=pt,
        network_name=net,
        background=bg,
        distance_type=dt
    )
    for pt, net, bg, dt in all_exps
]

runner_mod = "src.experiment_1.experiment_runner"
cwd = "/content/gestalt-DNNs"             # adjust to your repo root
env = {**os.environ, "PYTHONPATH": cwd}

for idx, params in enumerate(arguments, start=1):
    # dump params to a temp pickle
    with tempfile.NamedTemporaryFile(delete=False, suffix=".pkl") as tmp:
        pickle.dump(params, tmp)
        args_path = tmp.name

    print(f"\n▶ Experiment {idx}/{len(arguments)} → {params['network_name']} | {params['background']} | {params['distance_type']}")

    result = subprocess.run(
        [
            "python", "-m", runner_mod,
            "--args_path", args_path,
            "--batch_size", str(batch_size)
        ],
        cwd=cwd,
        env=env,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    print(result.stdout)
    if result.returncode != 0:
        print("❌ Subprocess failed:")
        print(result.stderr)

    os.remove(args_path)
```

```python
import numpy as np
import pickle
from itertools import product
from src.utils.create_stimuli.drawing_utils import DrawShape
from src.experiment_2.distance_helper import generate_stimuli_pairs

# Configurable options for data generation
file_path = "./experiment_2_data"
os.makedirs(file_path, exist_ok=True)

backgrounds = ['random'] #['random', 'white', 'black']
type_ds_options = [f'array{i}' for i in range(1, 19)]
type_ds_options += ['arrayF', 'arrayA', 'arrayB', 'arrayC', 'arrayD', 'arrayE']
type_ds_options += ['curly_composite_with_space', 'array4_curly', 'array11_curly']
transf_codes = ['t'] #['none', 't', 's', 'r']  # All transformation codes you might use

N = 500  # Number of samples per setting
img_size = np.array((224, 224), dtype=int)

for background, type_ds, transf_code in product(backgrounds, type_ds_options, transf_codes):
    draw_obj = DrawShape(
        background='black' if background in ['black', 'random'] else background,
        img_size=img_size,
        width=10
    )
    stimuli = generate_stimuli_pairs(
        draw_obj=draw_obj,
        type_ds=type_ds,
        N=N,
        var_tr=transf_code,
        type_ds_args=None  # Set if needed
    )
    filename = f'stimuli_{type_ds}_bg-{background}_tr-{transf_code}.pkl'
    with open(os.path.join(file_path, filename), 'wb') as f:
        pickle.dump(stimuli, f)
    print(f'Saved {filename}')
```

```python
#%%
import os
import subprocess
from itertools import product
from src.utils.misc import additional_nets


pretraining = ['ImageNet']
background = ['random']
distance_type = ['euclidean', 'cossim']
type_ds = [f'array{i}' for i in range(1, 19)]
type_ds += ['arrayF', 'arrayA', 'arrayB', 'arrayC', 'arrayD', 'arrayE']
type_ds += ['curly_composite_with_space', 'array4_curly', 'array11_curly']
transf_code = ['t']
nets = [
    'convnext_base',
    'regnet_y_16gf',
    'mobilenet_v3_large',
    'swin_tiny_patch4_window7_224',
    'deit_small_patch16_224',
    'coatnet_0_rw_224',
    'pit_s_224'
]

runner_script = '/content/gestalt-DNNs/src/experiment_2/experiment_runner2.py'  # Adjust path if needed
stimuli_dir = '/content/gestalt-DNNs/experiment_2_data/'  # Directory where stimuli_*.pkl files are stored

all_exps = product(pretraining, nets, type_ds, background, distance_type, transf_code)
arguments = [
    dict(pretraining=pt, network_name=net, type_ds=td, background=bg, distance_type=dt, transf_code=tc)
    for pt, net, td, bg, dt, tc in all_exps
]
batch_size = 16

for idx, args in enumerate(arguments):
    stimuli_file = f"stimuli_{args['type_ds']}_bg-{args['background']}_tr-{args['transf_code']}.pkl"
    stimuli_path = os.path.join(stimuli_dir, stimuli_file)
    print(f"\n▶ Running experiment {idx+1}/{len(arguments)}: {args['network_name']} {args['type_ds']} {args['distance_type']}")
    if not os.path.exists(stimuli_path):
        print(f"Stimuli file not found: {stimuli_path}")
        continue



    result = subprocess.run(
        [
            "python", runner_script,
            "--stimuli_path", stimuli_path,
            "--network_name", args['network_name'],
            "--pretraining", args['pretraining'],
            "--background", args['background'],
            "--distance_type", args['distance_type'],
            "--transf_code", args['transf_code'],
            "--type_ds", args['type_ds'],
            "--batch_size", str(batch_size)
        ],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True,
        cwd="/content/gestalt-DNNs",
        env={**os.environ, "PYTHONPATH": "/content/gestalt-DNNs"}
    )

    print(result.stdout)
    if result.returncode != 0:
        print("Subprocess failed with error:")
        print(result.stderr)
```
